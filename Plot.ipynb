# Scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np
x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,100,86,103,87,94,78,77,85,86]
plt.scatter(x,y,c='red',cmap='viridis')
plt.colorbar()
plt.show()
```

# Bubble plot

```python
import matplotlib.pyplot as plt
import numpy as pd
#create data
x=np.random.rand(40)
y=np.random.rand(40)
z=np.random.rand(40)
colors=np.random.rand(40)
#use the scatter plot
plt.scatter(x,y,s=z*1000,c=colors,alpha=0.6)
plt.show()
```
