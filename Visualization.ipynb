```python
import matplotlib.pyplot as plt
country=["USA","INDIA","Germany"]
GDP_PRE_CAPITA=[48000,38000,50000]
plt.bar(country,GDP_PRE_CAPITA)
plt.title("COUNTRY,GDP")
plt.xlabel("country")
plt.ylabel("GDP_PRE_CAPITA")
plt.show()
```
