# Pie chart

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
x=["r","b","hotpink"]
label=["DSP","CD","WT"]
plt.pie(y,colors=x,labels=label)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
x=["r","b","hotpink"]
plt.pie(y,colors=x,labels=["DSP","CD","WT"])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
x=["r","b","hotpink"]
plt.pie(y,colors=x,labels=["DSP","CD","WT"])
plt.legend(title="Subjects")
plt.show()
```

# Bar Graph

```python
import matplotlib.pyplot as plt
country=["USA","India","Germany","UK","France"]
GDP_per_capita=[48000,38000,50000,45000,44000]
plt.bar(country,GDP_per_capita,color="pink")
plt.title('country vs GDP per capita')
plt.xlabel('country')
plt.ylabel('GDP per capita')
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
numbers=[0.1,0.5,1,1.5,2,4,5.5,6,8,9]
plt.hist(numbers,bins=3,color="hotpink")
plt.xlabel("Number")
plt.ylabel("Frequency")
plt.show()
```
